```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#1

def generate_adjacency_matrix(filename):
  # write your code here
  infile = open(filename, 'r')
  lines = [line.strip() for line in infile]
  c = 0
  member = []
  for i in range(len(lines)) :
    lines[i] = lines[i].split(',')
    for name in lines[i] :
      if name not in member :
        c += 1
        member.append(name)
  A = np.zeros((c,c),int)
  for i in range(len(lines)) : 
    r = member.index(lines[i][0])
    c = member.index(lines[i][1])
    A[r,c] = 1
  person_names = member

  return A, person_names
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#2

def get_degree_matrix(A):
  # write your code here
  D = np.zeros((A.shape[0],A.shape[1]),int)
  for i in range(A.shape[0]) :
    count = 0
    for j in range(A.shape[1]) :
      if A[i,j] == 1 :
        count += 1
    D[i,i] = count
  
  return D
```

```python
D = get_degree_matrix(A)
print(D)
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#3

def get_names_with_highest_number_of_friends(D, person_names):
  # write your code here
  value = []
  for i in range(D.shape[0]) :
    for j in range(D.shape[1]) :
      if D[i,j] not in value :
        value.append(D[i,j])
  sortvalue = sorted(value)
  all_max = []
  for i in range(D.shape[0]) :
    for j in range(D.shape[1]) :
      if D[i,j] == sortvalue[-1] :
        all_max.append([person_names[i],i])
  all_max = [tuple(e) for e in all_max]     

  return all_max
```

```python
persons = get_names_with_highest_number_of_friends(D, person_names)
print(persons)
```

```python
# HELPER FUNCTION
def get_liked_page_matrix(infile):
  liked_pages = []
  persons = []
  fr = open(infile, 'r')
  for line in fr:
    name, bits = line.strip().split(',')
    liked_page = [int(c) for c in bits]
    liked_pages.append(liked_page)
    persons.append(name)
  liked_pages_array = np.array(liked_pages)
  return liked_pages_array,persons
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#4

def get_distance_matrix(Lp):
  # write your code here
  s = Lp.shape[0]
  distance_matrix = np.zeros((s,s),float)
  for i in range(distance_matrix.shape[0]) :
    for j in range(distance_matrix.shape[1]) :
      # compare rows = Lp[i,:] and Lp[j,:]
      row1 = Lp[i,:]
      row2 = Lp[j,:]
      a = 0
      for value in row1 :
        if value == 1 :
          a += 1
      b = 0
      for value in row2 :
        if value == 1 :
          b += 1
      #find c 
      c = 0
      for x in range(Lp.shape[1]) :
        if row1[x] == 1 and row1[x] == row2[x] :
          c += 1
      tanito = c / (a+b-c)
      if i != j :
        distance_matrix[i,j] = tanito
    
  return distance_matrix
get_distance_matrix(liked_pages)
  
```

```python
Dt = get_distance_matrix(liked_pages)
print(Dt)
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#5

def get_all_most_similar_pairs(Dt, persons):
  # write your code here
  value = []
  for i in range(Dt.shape[0]) :
    for j in range(Dt.shape[1]) :
      if Dt[i,j] not in value :
        value.append(Dt[i,j])
  #print(sorted(value))

  p = []
  for i in range(Dt.shape[0]) :
    for j in range(Dt.shape[1]) :
      if Dt[i,j] == sorted(value)[-1] :
        p.append(sorted([persons[j],persons[i]]))

  setp = []
  for elem in p :
    if elem not in setp:
      setp.append(elem) 
  pairs = [tuple(e) for e in setp]
  
  return pairs

```

```python
pairs = get_all_most_similar_pairs(Dt, persons)
print(pairs)
```
