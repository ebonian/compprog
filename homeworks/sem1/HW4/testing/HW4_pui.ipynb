```python
import numpy as np
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#1

def generate_adjacency_matrix(filename):
  # write your code here
  file = open(filename)
  person_names = [] 
  d = {}
  for lines in file:
    a,b = lines.strip().split(',')
    if a not in d:
      d[a] = b
    else:
      d[a] += b
  for names in d:
    if names not in person_names:
      person_names.append(names)
  
  A = np.zeros((len(person_names), len(person_names)), int)
  for r in range(len(person_names)):
    for c in range(len(person_names)):
      if person_names[c] in d[person_names[r]]:
        A[r,c] = 1  

  return A, person_names

```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#2

def get_degree_matrix(A):
  # write your code here
  D = np.zeros((A.shape[0], A.shape[1]), int)
  n = np.sum([A], axis=1)
  D[np.arange(A.shape[0]),np.arange(A.shape[1])] = n
  
  return D
```

```python
D = get_degree_matrix(A)
print(D)
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#3

def get_names_with_highest_number_of_friends(D, person_names):
  # write your code here
  all_max = []
  max = np.max(D)
  n = np.max(D, axis=1)
  for i in range(len(n)):
    if n[i] == max:
      all_max.append((person_names[i], i))

  return all_max
```

```python
persons = get_names_with_highest_number_of_friends(D, person_names)
print(persons)
```

```python
# HELPER FUNCTION
def get_liked_page_matrix(infile):
  liked_pages = []
  persons = []
  fr = open(infile, 'r')
  for line in fr:
    name, bits = line.strip().split(',')
    liked_page = [int(c) for c in bits]
    liked_pages.append(liked_page)
    persons.append(name)
  liked_pages_array = np.array(liked_pages)
  return liked_pages_array,persons
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#4

def get_distance_matrix(Lp):
  # write your code here
  distance_matrix = np.zeros((Lp.shape[0], Lp.shape[0]))
  for r in range(-1, Lp.shape[0]):
    for r1 in range(-1, Lp.shape[0]):
      a = np.sum(Lp, axis=1)[r]
      b = np.sum(Lp, axis=1)[r1]
      c = np.sum(Lp[r,:]*Lp[r1,:])
      tc = c/(a+b-c)
      if r == r1:
        distance_matrix[r,r1] = float(0)
      else:
        distance_matrix[r,r1] = tc
      
  return distance_matrix
```

```python
Dt = get_distance_matrix(liked_pages)
print(Dt)
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#5

def get_all_most_similar_pairs(Dt, persons):
  # write your code here
  pairs = []
  highest = np.max(Dt)
  for r in range(Dt.shape[0]):
    for c in range(Dt.shape[1]):
      if  Dt[r,c] == highest:
        if r < c:
          pairs.append((persons[r], persons[c]))

  return pairs
```

```python
pairs = get_all_most_similar_pairs(Dt, persons)
print(pairs)
```
