```python
import numpy as np
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#1

def generate_adjacency_matrix(filename):
    # write your code here
    with open(filename, 'r') as f:
        lines = np.array(f.readlines())

    def _a(line):
        line = line.strip().split(',')
        return line
    rel = np.array([_a(line) for line in lines])
    # print(rel)

    keys = sorted(list(set(rel.flatten())))

    d = {k: i for i, k in enumerate(keys)}

    m = np.zeros((len(keys), len(keys)), dtype=int)

    rel2 = np.array([[d[r[0]], d[r[1]]] for r in rel])

    r, c = zip(*rel2)
    m[r, c] = 1
    m[c, r] = 1

    A = m
    person_names = keys




    return A, person_names
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#2

def get_degree_matrix(A):
    # write your code here
    A = np.sum(A, axis=0)
    A = np.diag(A)
    D = A

    return D
```

```python
D = get_degree_matrix(A)
print(D)
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#3

def get_names_with_highest_number_of_friends(D, person_names):
    # write your code here
    D = np.sum(D, axis=0)
    p = np.where(D == np.max(D))
    all_max = [tuple([person_names[i], i]) for i in p[0]]

    return all_max
```

```python
persons = get_names_with_highest_number_of_friends(D, person_names)
print(persons)
```

```python
# HELPER FUNCTION
def get_liked_page_matrix(infile):
  liked_pages = []
  persons = []
  fr = open(infile, 'r')
  for line in fr:
    name, bits = line.strip().split(',')
    liked_page = [int(c) for c in bits]
    liked_pages.append(liked_page)
    persons.append(name)
  liked_pages_array = np.array(liked_pages)
  return liked_pages_array,persons
```

```python
"""
from google.colab import files
uploaded = files.upload()
print('uploaded = ', uploaded)
"""
liked_pages, persons = get_liked_page_matrix('liked_pages.txt')
print(liked_pages)
print(persons)
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#4

def get_distance_matrix(Lp):
    # write your code here
    m = np.array(Lp)
    keys = np.array(persons)

    o = (len(keys), len(keys))

    def _b(r, c):
        return np.sum(m[r] & m[c]) / np.sum(m[r] | m[c])
    output = [_b(r, c) for r, c in np.ndindex(o)]
    output = np.array(output).reshape(o)
    np.fill_diagonal(output, 0)
    distance_matrix = output


    return distance_matrix
```

```python
Dt = get_distance_matrix(liked_pages)
print(Dt)
```

```python
# DO NOTE DELETE THIS CELL / WRITE YOUR CODE HERE
# Task#5

def get_all_most_similar_pairs(Dt, persons):
    # write your code here
    Dt = np.triu(Dt)
    p = np.where(Dt == np.max(Dt))
    if len(p[0]) == 1:
        p = [[p[0][0], p[1][0]]]
    p = np.array([(persons[x[0]], persons[x[1]]) for x in p])
    pairs = [tuple(sorted(x)) for x in p]
    
    return pairs
```

```python
pairs = get_all_most_similar_pairs(Dt, persons)
print(pairs)
```
